Jupyter Notebook to create a list of values and the number of cycles needed for the value to be transformed to the integer 1.

The Collatz conjecture is related to a sequence that is defined on Wikipedia as "start with any positive integer $n$. Then each term is obtained from the previous term as follows: if the previous term is even, the next term is one half the previous term. If the previous term is odd, the next term is 3 times the previous term plus 1. The conjecture is that no matter what value of $n$, the sequence will always reach 1."

## Import Modules

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 10] # change plot size [width, height] --in inches
```

## Input Sample Size

```python
# user-input for sample size

n = int(input('> '))
```

## Create Dataframe & Populate with Values

```python
# create empty pandas dataframe to fill with data

df = pd.DataFrame(columns=['initialValue', 'stoppingTime'])

# loop through values in range of sample size, starting with i = 2

for i in range(2,n):
    
    stoppingTime = 0
    j = i

    # apply collatz conjecture rules

    while j != 1:
        if j % 2 == 0:
            j = j / 2
            stoppingTime += 1
        else:
            j = 3*j + 1
            stoppingTime += 1

    # add starting value, number of steps until j = 1 to dataframe

    df = df.append(pd.DataFrame({'initialValue': i, 'stoppingTime': stoppingTime}, index=[0]), ignore_index=True)
```

## Plot the Starting Value _vs._ Stopping Time

```python
df.plot(x='initialValue', y='stoppingTime')
```
